# <font color='yellow'> Matplotlib Plots

### <font color='pink'> Ref: https://matplotlib.org/stable/

Matplotlib is a plotting library in Python, allowing to create a wide range of static, animated, and interactive plots.

1. **Line Plot** (`plt.plot()`): 
   - Displays data points as connected lines.
   - Used for time series or continuous data.
   
   ```python
   plt.plot(x, y)
   ```

2. **Scatter Plot** (`plt.scatter()`):
   - Displays data points as individual dots.
   - Useful for showing correlations between two continuous variables.
   
   ```python
   plt.scatter(x, y)
   ```

3. **Bar Plot** (`plt.bar()`):
   - Displays categorical data with rectangular bars.
   - Useful for comparing the size of different categories.
   
   ```python
   plt.bar(x, y)
   ```

4. **Horizontal Bar Plot** (`plt.barh()`):
   - Similar to the bar plot, but bars are horizontal.
   
   ```python
   plt.barh(x, y)
   ```

5. **Histogram** (`plt.hist()`):
   - Displays the distribution of a dataset.
   - Useful for understanding the frequency distribution of a variable.
   
   ```python
   plt.hist(data, bins=20)
   ```

6. **Box Plot** (`plt.boxplot()`):
   - Displays the distribution of data through quartiles.
   - Highlights the median, outliers, and interquartile range.
   
   ```python
   plt.boxplot(data)
   ```

7. **Pie Chart** (`plt.pie()`):
   - Displays proportions of categories in a circular chart.
   
   ```python
   plt.pie(sizes, labels=labels)
   ```

8. **Balance Plot** (`plt.fill_between()`):
   - Displays a cumulative sum or Balance beneath a line plot.
   
   ```python
   plt.fill_between(x, y)
   ```

9. **Violin Plot** (`plt.violinplot()`):
   - Similar to a box plot but shows the probability density of data at different values.
   
   ```python
   plt.violinplot(data)
   ```

10. **Hexbin Plot** (`plt.hexbin()`):
    - Used to visualize the relationship between two continuous variables through hexagonal bins.
    
    ```python
    plt.hexbin(x, y, gridsize=30)
    ```

11. **Heatmap** (`plt.imshow()`):
    - Displays 2D data in matrix form with color coding for values.
    
    ```python
    plt.imshow(data, cmap='hot', interpolation='nearest')
    ```

12. **Error Bar Plot** (`plt.errorbar()`):
    - Displays data points with error bars to show uncertainty or variability.
    
    ```python
    plt.errorbar(x, y, yerr=errors)
    ```

13. **Stacked Plot** (`plt.stackplot()`):
    - Creates a stacked Balance plot to show how multiple variables contribute to a total over time.
    
    ```python
    plt.stackplot(x, y1, y2, y3)
    ```

14. **Polar Plot** (`plt.polar()`):
    - Plots data in polar coordinates (angles and distances).
    
    ```python
    plt.polar(theta, r)
    ```

15. **Step Plot** (`plt.step()`):
    - Similar to a line plot but with steps between data points, useful for discrete data.

    ```python
    plt.step(x, y)
    ```

# <font color='yellow'> Seaborn Plots

### <font color='pink'> Ref: https://seaborn.pydata.org/index.html#

Seaborn is built on top of Matplotlib and provides a higher-level interface for creating attractive and informative statistical graphics.

1. **Scatter Plot** (`sns.scatterplot()`):
   - Similar to Matplotlib's scatter plot, but with additional features like automatic color mapping.
   
   ```python
   sns.scatterplot(x='x_column', y='y_column', data=df)
   ```

2. **Line Plot** (`sns.lineplot()`):
   - Displays a line plot, and it can also handle confidence intervals.
   
   ```python
   sns.lineplot(x='x_column', y='y_column', data=df)
   ```

3. **Bar Plot** (`sns.barplot()`):
   - A bar plot that can summarize data using the mean or other statistics.
   
   ```python
   sns.barplot(x='x_column', y='y_column', data=df)
   ```

4. **Count Plot** (`sns.countplot()`):
   - Displays counts of observations in categorical data.
   
   ```python
   sns.countplot(x='category_column', data=df)
   ```

5. **Box Plot** (`sns.boxplot()`):
   - Similar to Matplotlib's box plot but with more functionality for visualizing distributions.
   
   ```python
   sns.boxplot(x='category_column', y='value_column', data=df)
   ```

6. **Violin Plot** (`sns.violinplot()`):
   - Combines aspects of box plots and density plots to visualize data distribution.
   
   ```python
   sns.violinplot(x='category_column', y='value_column', data=df)
   ```

7. **Heatmap** (`sns.heatmap()`):
   - Visualizes matrix-like data with color-coded values.
   
   ```python
   sns.heatmap(data)
   ```

8. **Pair Plot** (`sns.pairplot()`):
   - Visualizes pairwise relationships in a dataset with scatter plots and histograms on the diagonal.
   
   ```python
   sns.pairplot(df)
   ```

9. **Facet Grid** (`sns.FacetGrid()`):
   - Allows you to create a grid of plots, which can be useful for comparing multiple categories or groups.
   
   ```python
   g = sns.FacetGrid(df, col="category_column")
   g.map(sns.histplot, "value_column")
   ```

10. **Joint Plot** (`sns.jointplot()`):
    - Displays bivariate data and includes both a scatter plot and marginal histograms.
    
    ```python
    sns.jointplot(x='x_column', y='y_column', data=df)
    ```

11. **Dist Plot** (`sns.histplot()`):
    - Displays a histogram with an optional kernel density estimate (KDE) overlay.
    
    ```python
    sns.histplot(df['column'])
    ```

12. **Strip Plot** (`sns.stripplot()`):
    - Displays individual data points along a categorical axis.
    
    ```python
    sns.stripplot(x='category_column', y='value_column', data=df)
    ```

13. **Swarm Plot** (`sns.swarmplot()`):
    - Similar to a strip plot, but ensures that points do not overlap.
    
    ```python
    sns.swarmplot(x='category_column', y='value_column', data=df)
    ```

14. **Regression Plot** (`sns.regplot()`):
    - A scatter plot with an optional regression line fitted to the data.
    
    ```python
    sns.regplot(x='x_column', y='y_column', data=df)
    ```

15. **Kde Plot** (`sns.kdeplot()`):
    - Plots the kernel density estimate of a distribution.
    
    ```python
    sns.kdeplot(df['column'])
    ```

16. **Balance Plot** (`sns.Balance_plot()`): 
    - Seaborn doesn't have a direct method for Balance plots, but you can use `sns.lineplot()` to create an Balance plot by filling between the lines.
    
    ```python
    sns.lineplot(x='x_column', y='y_column', data=df, fill=True)
    ```
